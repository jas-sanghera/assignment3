```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as LA
%matplotlib qt
from scipy.stats import mode
rng = np.random.default_rng()

#function to pick a value for 'b' that is not the same value as given 'a'
#if 'b' is the same as 'a', then it picks a new 'b'
def pickb(a, hat):
    b = rng.integers(0, len(hat))
    while a == b:
        b = rng.integers(0, len(hat))
    return b

#function to remove the a-th and b-th value from 'hat'
#must delete the higher index first
def deleteab(a,b,hat):
    if a > b:
        hat = np.delete(hat, a)
        hat = np.delete(hat, b)
    else:
        hat = np.delete(hat, b)
        hat = np.delete(hat, a)
    return hat

#function which mimics the assignment problem
#removes 2 numbers from hat, calculates the difference, places new number in the hat, until 1 number remains
def hatpicking(year):
    hat = np.arange(1, year+1)
    for i in range(year-1):
        a = rng.integers(0, len(hat))
        b = pickb(a, hat)
        if hat[a] > hat[b]:
            c = hat[a]- hat[b]
        else:
            c = hat[b] - hat[a]
        hat = np.append(hat, c)
        hat = deleteab(a,b,hat)
    return hat

#running the scenario to collect 'samples'
samples = 10000
X = np.zeros(samples)
for i in range(samples):
    X[i] = hatpicking(2024)

#plotting a histogram
Max = max(X)
Min = min(X)
bins = np.arange(Min, Max+2)-0.5
fig, ax = plt.subplots(1,1)
Data, B1, B2 = ax.hist(X, bins, density = True, rwidth=0.8)
ax.set_xlim(Min -1, Max+1)
ax.set_xlabel('X')
ax.set_ylabel('Probability Mass Function')
plt.grid('on')
plt.show

#calulating the mode and the maximum from the samples
Mode = mode(X)
Max

#calculating the proportion of values which are even
c = 0
for i in range(len(X)):
    if  X[i] % 2 == 0: #can be changed to different criteria e.g if  X[i] < 100:
        c += 1
c/samples
```
